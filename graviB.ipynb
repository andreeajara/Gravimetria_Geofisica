```python
data = loadtxt("gravib.csv", delimiter=',')
grav = data[:,0]
hora = data[:,1]*60
minuto = data[:,2]
seg = data[:,3]/60
elev = data[:,4]
#densidad neis
dn=2800 #kg/m³
```

```python
time = hora+minuto+seg
```

```python
scatter(time,grav)
xlabel('tiempo (min)')
ylabel('gravedad')
title('tiempo vs. gravedad observada')
```
