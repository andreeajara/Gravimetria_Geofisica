```python
data = loadtxt("gravig.csv", delimiter=',')
grav = data[:,0]
hora = data[:,1]*60
minuto = data[:,2]
seg = data[:,3]/60
elev2 = data[:,4]
#densidad esquistos
ds = 2460 #kg/m³
```

```python
elev3 =  []
for i in range(size(elev2)):
    elev3.append(elev2[i])
elev = list(reversed(elev3))
```

```python
time = hora+minuto+seg
```

```python
scatter(time,grav)
xlabel('tiempo (min)')
ylabel('gravedad')
title('tiempo vs. gravedad observada')
```

Correccion de deriva

```python
m1 = (grav[size(grav)-1]-(grav[0]))/(time[size(time)-1]-time[0])
m2 = (grav[size(grav)-2]-(grav[1]))/(time[size(time)-2]-time[1])
```

```python
gdrift = []

for i in range((size(grav))):
    gdrift.append(grav[i]-(m1*(time[i]-time[0])))
```

```python
scatter(time,gdrift)
xlabel('tiempo(min)')
ylabel('gravedad')
title('gravedad corregida drift')
```

```python
esarmero = []
for i in range(size(elev)):
    esarmero.append(elev[0]-elev[i])
```

```python
gFA = []
for i in range(len(gdrift)):
    gFA.append(gdrift[i]-(esarmero[i]*0.3086))
```

```python
scatter(elev,gFA)
xlabel('elevacion(m)')
ylabel('gravedad')
title('gravedad corregida drift+free air')
```
